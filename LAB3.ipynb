## Exp3: Data processing and cleaning

```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
display(df.head())
```

```python
df.isnull()
```

```python
import numpy as np

for i in range(4):
  df.loc[i, 'sepal length (cm)'] = np.nan
for i in range(7,11):
  df.loc[i, 'petal length (cm)'] = np.nan
for i in range(4,7):
  df.loc[i, 'sepal width (cm)'] = np.nan
for i in range(7,11):
  df.loc[i, 'petal width (cm)'] = np.nan
```

```python
df.head(10)
```

```python
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        mean_value = df[col].mean()
        df[col].fillna(mean_value, inplace=True)

display(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
df.boxplot()
plt.title("Box Plot of Iris Dataset (Missing Values Treated)")
plt.ylabel("Measurement (cm)")
plt.show()
```

# Box Plot for 8 values

```python
import matplotlib.pyplot as plt
import numpy as np

arr1=[100,120,130,150,180,200,220,250]

plt.figure(figsize=(6, 4))
plt.boxplot(arr1)
plt.title("Box Plot of 8 values")
plt.ylabel("Values")
plt.show()
```

```python

q1 = np.percentile(array1, 25)
median = np.percentile(array1, 50)
q3 = np.percentile(array1, 75)

print(f"25th percentile (Q1): {q1}")
print(f"50th percentile (Median): {median}")
print(f"75th percentile (Q3): {q3}")
```
